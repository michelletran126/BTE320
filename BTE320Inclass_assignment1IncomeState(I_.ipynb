```python
sales = float(input('Enter sales ($): '))
print(f"Sales: ${sales}")
tax = float(input('Enter tax rate (%): '))
print(f"Tax Rate: ${tax}")

#Cost of Goods Sold
cog = 0.4
cog_sold = sales * cog
print(f"Cog: ${cog_sold}")


#Gross Profit
grossProfit = sales - (cog*sales)
print(f"Gross profit: ${grossProfit}")

#Taxes
taxes = grossProfit * (tax/100)
print(f"Taxes: ${taxes}")

netIncome = grossProfit - taxes
print(f"Net Income ${netIncome}")


```
