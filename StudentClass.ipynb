```python
class Student:
  def __init__(self, name, number):
    self.__name = name
    self.__number = number
    self.__courses = []

  def get_name(self):
    return self.__name

  def get_courses(self):
    return self.__courses

  def set_name(self, newname):
    self.__name = newname

  def __len__(self):
    return len(self.__courses)

  def enroll(self, course):
    if course in self.__courses:
      print('You have already enrolled in', course)
    else:
      self.__courses.append(course)

s = Student('Bob', 12345)
s.get_name()
s.get_courses()
s.set_name('Bob Andrews')
s.get_name()
s.enroll('ABC123')
s.enroll('ABC123')
s.enroll('ABC567')
len(s)
s.get_courses()
```
