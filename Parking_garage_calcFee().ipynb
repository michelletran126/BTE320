```python
def hours(h):
  fee = 5 + 2.5 * h
  if fee < 10:
    return 10.0
  elif fee > 20:
    return 20.0
  else:
    return fee

def main():
  user_hours = int(input('Enter number of hours for parking: '))
  fee = hours(user_hours)
  print(fee)


if __name__ == "__main__":
  main()

Recursive:
def reverse(s):
  if len(s) > 1:
    return s[-1] + reverse(s[1:-1]) + s[0]
  else:
    return s

reverse('hello')
```
