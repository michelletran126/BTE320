```python

for hours in range(1, 9):
   fee = 5 + hours * 2.5
   if fee < 10:
    fee = 10.0
    print(f"{hours}, {fee}")
   elif fee > 20:
    fee = 20.0
    print(f"{hours}, {fee}")
   else:
    print(f'{hours}, {fee}')
```
