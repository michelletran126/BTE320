```python
def reverse(s):
    # Base case
    if s == "":
        return ""
    # Recursive step
    return reverse(s[1:]) + s[0]

# Examples
print(reverse("hello"))   # → 'olleh'
print(reverse("12345"))   # → '54321'
```
