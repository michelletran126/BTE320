```python
def loan_payment(principal: float, annual_rate: float, months: int) -> float:
  assert isinstance(principal, (int, float)), "Principal must be numerical"
  assert isinstance(annual_rate, (int, float)), "annual_rate must be numerical"
  assert isinstance(months, int), "months must be an integer"

  if principal <= 0:
    raise ValueError("Principal must be greater than 0")
  if annual_rate <= 0:
    raise ValueError("Annual rate cannot be negative")
  if months <= 0:
    raise ValueError("Months cannot be negative")

  r = annual_rate / 12.0
  n = months
  P =float(principal)

  if r == 0.0:
   return P/n
  else:
    return r * P / (1.0 - (1.0 + r) ** (-n))

if __name__ == "__main__":
  try:
    P = float(input("Enter the principal amount: "))
    apr = float(input("Enter the annual interest rate: "))
    n = int(input("Enter the number of months: "))
    payment = loan_payment(P, apr, n)
    print(f"The monthly payment is ${payment:.2f}")
  except Exception as e:
    print(e)
```
