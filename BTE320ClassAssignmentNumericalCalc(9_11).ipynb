```python
num1 = float(input("First number: "))
num2 = float(input("Second number: "))
op = input("Choose numberical operator (+, -, *, /, //, **, %): ")

if op == '+':
  print(f"{num1} + {num2} =  {num1 + num2}")
elif op == '-':
  print(f"{num1} - {num2} = {num1 - num2}")
elif op == '*':
  print(f"{num1} * {num2} == {num1 * num2}")
elif op == '/':
  print(f"{num1} / {num2} == {num1 / num2}")
elif op == '//':
  print(f"{num1} // {num2} == {num1 // num2}")
elif op == '**':
  print(f"{num1} ** {num2} == {num1 ** num2}")
elif op == '%':
  print(f"{num1} mod {num2} == {num1 % num2}")
else:
  print("Invalid operator")
```
