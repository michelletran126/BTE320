````python
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams.update({'font.size': 20, 'figure.figsize': (10, 8)})
````

````python
df = pd.read_csv("travel-times.csv")
````

````python
#print the shape of the data set

df.info()
df.shape
````

````python
# print the first 15 rows of the data set

df.head(15)
````

````python
# get information for the features (column) with missing values

df.isnull()
````

````python
#Drop duplicate values (if any) by keeping only the last instance

df.drop_duplicates(inplace=True, keep='last')

````

````python
#calculate the total number of missing values(if any) on each column

df.isnull().sum()
````

````python
# create two copies of the dataframe, and then:
# drop rows with missing values from the 1st copy
# drop columns with missing values from the 2nd copy

# Create two copies of the dataframe
df_rows_dropped = df.copy()
df_cols_dropped = df.copy()

# Drop rows with missing values from the first copy
df_rows_dropped = df_rows_dropped.dropna()

# Drop columns with missing values from the second copy
df_cols_dropped = df_cols_dropped.dropna(axis=1)
````

````python
# get summary statistics and see the correlation between the numerical columns
print("Summary Statistics:")
df.describe()
````

````python
# shows rows 11 to 14

df.iloc[11:15]
````

````python
# Create subset for the two specific dates
subset = df[(df['Date'] == '2011-11-23') | (df['Date'] == '2012-01-06')]

subset
````

````python
#produce a scatterplot between Distance and TotalTime (1 points)
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20, 'figure.figsize': (10, 8)})
df.plot(kind='scatter', x='Distance', y='TotalTime', title='Distance vs TotalTime')
````

````python
#produce boxplots for AvgSpeed and AvgMovingSpeed (use different cells for each)

df['AvgSpeed'].plot(kind='box', title='Boxplot of AvgSpeed')
plt.show()
df['AvgMovingSpeed'].plot(kind='box', title='Boxplot of AvgMovingSpeed')
plt.show()
````
