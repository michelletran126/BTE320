```python
Append items to list iteratively (while loop):
●
Starting with an empty list, design a program in Python that populates the list by
repeatedly asking the user for numerical inputs.
●
The process stops when the user gives 0.
●
Finally, the list should be displayed in descending order.
```

```python
num = int(input('Enter a number: '))
list = []

while num > 0:
  list.append(num)
  print(list)
  num = int(input('Enter a number (0 to exit): '))

list.sort(reverse=True)

print('Sorted list in descending order: ', list)
```
