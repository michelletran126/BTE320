```python
products = {
    'P-01': 9.95,
    'P-02':10.50
}

id = input('Enter Product ID: ')

if id in products:
  print(f'Product ID: {id}, Price: ${products[id]}')
else:
  print(f'Product ID {id} not found')


```

```python
## Dynamic dictionary; N distinct key:value pairs(while loop)/ if elif-else statements)
## Populate an empty dictionary interatively N times (choose your own value for N) to add name:count pairs.
##mThe key must be given as input from the user on every interation. Use conditionals to test if the key has
## been used before. If not, set count to 1; otherwise, increase count by 1


names = {}
\
for i in range(5):
  name = input('Enter a name: ')

  if name in names:
    names[name] = names[name] + 1 # names[name] += 1
  else:
    names[name] = 1

  print(names)
```
